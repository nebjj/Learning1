```python
def life_in_weeks(age):
    weeks_remaining = 52*(90-age)
    print(f"You have {weeks_remaining} weeks left.")

life_in_weeks(83)
```

```python
def calculate_love_score(name1, name2):
    names=name1+name2
    t=names.count("t")
    r=names.count("r")
    u=names.count("u")
    e=names.count("e")
    l=names.count("l")
    o=names.count("o")
    v=names.count("v")

    first_digit = t+r+u+e
    second_digit = l+o+v+e

    score = str(first_digit)+str(second_digit)

    return int(score)

calculate_love_score("Nebic Johnson", "Jenifer Lopez")
```

```python
import string
alphabet = string.ascii_lowercase

to_encrypt_text = input("Enter the text to encrypt: ")

to_encrypt_text=to_encrypt_text.lower()

def encrypt(to_encrypt_text, shift_number):
  encrypt_word=""
  for letter in to_encrypt_text:
    if letter in alphabet:

      position=alphabet.index(letter)
      new_position = (position+shift_number) % len(alphabet)
      new_letter=alphabet[new_position]
      encrypt_word += new_letter
    else:
      encrypt_word += letter
  return encrypt_word

shift_number = int(input("Enter the 'shift number' for encryption: "))

encrypt_text = encrypt(to_encrypt_text, shift_number)

print(f"Encrypted text: {encrypt_text}")


def deencrypt(encrypt_text, shift_number):
  deencrypt_word=""
  for letter in encrypt_text:
    if letter in alphabet:

      position=alphabet.index(letter)
      new_position = (position-shift_number)
      new_letter=alphabet[new_position]
      deencrypt_word += new_letter
    else:
      deencrypt_word += letter
  return deencrypt_word

deencrypt_text = deencrypt(encrypt_text, shift_number)


print(f"Decrypted text: {deencrypt_text}")
```
