```python
nums = [1,1,2,2,3,4,4]
```

```python
for i in range(len(nums)):
  for j in range(i+1,len(nums)):
    print(i,j)
    if nums[i] == nums[j]:
      print(f"Duplicate found. It is {nums[i]}")
      print(nums)
```

```python
current_len = len(nums)
i = 0
while i < current_len:
  j = i + 1
  while j < current_len:
    print(i, j)
    if nums[i] == nums[j]:
      print(f"Duplicate found. It is {nums[i]}")
      nums.pop(j) # Use pop by index to remove and shift.
      current_len -= 1 # Decrement current_len as an element is removed
      print(nums)
    else:
      # Only increment j if no removal happened, otherwise same j needs re-check
      # because a new element has shifted into nums[j]'s position.
      j += 1
  i += 1
```

```python
nums = [1,1,2,2,3,4,4]
current_length = len(nums)
i=0

while i < current_length:
  j = i + 1
  while j < current_length:
    print(i,j)
    if nums[i] == nums[j]:
      nums.pop(j)
      current_length -= 1
      print(nums)
    else:
      j += 1
  i += 1
```
