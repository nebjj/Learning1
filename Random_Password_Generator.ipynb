```python
import string
letters_list=list(string.ascii_letters)
numbers_list=list(string.digits)
spl_characters_list=['@','#','$','%',"!","*","&"]
```

```python
print(letters_list)
```

```python
print(numbers_list)
```

```python
print(spl_characters_list)
```

```python
numbers=int(input("How many numbers would you like in the password?"))
```

```python
spl_chara=int(input("How many special characters would you like in the password?"))
```

```python
letters=int(input("How many letters would you like in the password?"))
```

```python
import random
random_letters = random.sample(letters_list, letters)
print(random_letters)
```

```python
random_numbers = random.sample(numbers_list,numbers)
print(random_numbers)
```

```python
random_characters=random.sample(spl_characters_list,spl_chara)
print(random_characters)
```

```python
password_characters=list(random_characters+random_letters+random_numbers)
password_length=letters+numbers+spl_chara
password_list=random.sample(password_characters,password_length)
```

```python
password = ''.join(password_list)
```

```python
print(password)
```
