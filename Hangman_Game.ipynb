```python
import random

word_list=['brush','tablet','toffee','phone','laptop','cup']

word=random.choice(word_list)
```

```python
word
```

```python
length=len(word)
```

```python
print(f"It's a {length} letter word")
```

```python
correct_letters=[]
```

```python
game_over = False
```

```python
while not game_over:
  guess = str(input("Enter a letter ")).lower()
  display = ""

  for each_letter in word:
    if each_letter == guess:
      display += each_letter
      correct_letters.append(guess)
    elif each_letter in correct_letters:
      display += each_letter
    else:
      display += " _ "

  print(display)

  if "_" not in display:
      game_over = True
      print("Edaaa Sunnyykuttaaa neee jaichedaaa!!")
```
