```python
def add(number1, number2):
  return number1 + number2

def subtract(number1, number2):
  return number1 - number2

def multiply(number1, number2):
  return number1 * number2

def divide(number1, number2):
  return number1 / number2


operation = {
    "+" : add,
    "-" : subtract,
    "*" : multiply,
    "/" : divide
}
```

```python
number1 = int(input("Enter first number: "))
for symbol in operation:
  print(symbol)
operator = input("Which is the operator would you like to use? ")
number2 = int(input("Enter second number: "))
answer = operation[operator](number1, number2)
print(f"{number1} {operator} {number2} = {answer}")
```

```python
def calculator():
  number1 = int(input("Enter first number: "))
  def continuinge():
    continuing = True
    nonlocal number1
    while continuing is not False:
      for symbol in operation:
        print(symbol)
      operator = input("Which is the operator would you like to use? ")
      number2 = int(input("Enter second number: "))
      answer = operation[operator](number1, number2)
      print(f"{number1} {operator} {number2} = {answer}")
      keep_doing = input("Are you continue calculating with the result? 'yes' or 'no': ")
      if keep_doing.lower() == "yes":
        number1 = answer
        return continuinge()
      if keep_doing.lower() == "no":
        then_continuing = input("Would you like to continue? 'yes' or 'no'? ")
        if then_continuing.lower() == "yes":
          return calculator()
        if then_continuing.lower() == "no":
          continuing = False
          print("Good Bye! I miss you. \n - Calculator")
  continuinge()


calculator()
```
