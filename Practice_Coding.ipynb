```python
#ask name
#ask bid
#make it a dictionary
#ask for next
#at last - person with highest bid gets it
```

```python
bidder_list={}

flag = True

while flag == True:
  name = input("Enter your name ")
  bid_amount = int(input("Enter your bid amount: ₹"))
  to_next = input("Is there any other bidders? Yes or No ")
  #print("\n"*100)
  bidder_list[name]=bid_amount

  if to_next.lower() == "no":
    flag = False

print(bidder_list)
```

```python
def highest_bidder(bid_list):
  highest_bidder_name = max(bid_list, key=bid_list.get)
  max_bid_amount = bid_list[highest_bidder_name]

  print(f"The highest bid amount is ₹{max_bid_amount} and the tender goes to {highest_bidder_name}")

highest_bidder(bidder_list)
```

```python
'''- on every year that is divisible by 4 with no remainder

- except every year that is evenly divisible by 100 with no remainder

- unless the year is also divisible by 400 with no remainder '''

def is_leap_year(year):

    if year % 4 == 0:
      if year % 100 == 0:
        if year % 400 == 0:
          return True
        else:
          return False
      else:
        return True
    else:
      return False

is_leap_year(1989)
```

```python
def calculator(operation,first_number, second_number):
  if operation == "add":
    return add(first_number,second_number)
  if operation == "subtract":
    return subtract(first_number,second_number)
  if operation == "multiply":
    return multiply(first_number,second_number)
  if operation == "divide":
    return divide(first_number,second_number)

def add(a,b):
  return a+b
def subtract(a,b):
  return a-b
def multiply(a,b):
  return a*b
def divide(a,b):
  return a//b

operator = input("Which operation would you like to do? add, subtract, multiply or divide: ").lower()
first_number = int(input("Enter the first number: "))
second_number = int(input("Enter the second number: "))
answer = calculator(operator, first_number, second_number)

print(f"The resultant of the operation between {first_number} and {second_number} is {answer}")
```
